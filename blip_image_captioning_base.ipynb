```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("image-text-to-text", model="Salesforce/blip-image-captioning-base")
```

```python
# Load model directly
from transformers import AutoProcessor, BlipForConditionalGeneration

processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
from PIL import Image
import requests

# Sample image URL for demonstration.
# Replace this with your uploaded image path if you have one, e.g., img_path = '/content/my_uploaded_image.jpg'
img_url = 'https://huggingface.co/datasets/hf-internal-testing/fixtures_doc_examples/resolve/main/split_of_views.png'

# Open the image from URL or local path
try:
    image = Image.open(requests.get(img_url, stream=True).raw)
except Exception as e:
    print(f"Could not open image from URL: {e}")
    # If you have an uploaded image, you can try loading it like this:
    # image = Image.open('/content/your_uploaded_image.jpg')
    # Make sure to uncomment and update the path accordingly.
    exit()

# Generate caption using the pipeline
caption = pipe(image)

print(f"Generated Caption: {caption}")
```
